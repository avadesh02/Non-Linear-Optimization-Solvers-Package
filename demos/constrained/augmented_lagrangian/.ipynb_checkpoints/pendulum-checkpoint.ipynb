```python
# This file is a demo of using augmented lagrangian solver to compute a feasible trajectory for the
# pendulum 
import numpy as np
import IPython
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import FuncAnimation

import torch
from py_solvers.constrained.augmented_lagrangian.bound_constrained import BoundConstrainedFormulation
```

```python
class PendulumEnv:
    
    def __init__(self, l, m, dt):
        '''
        Input:
            l : lenght of the rod
            m : mass
            dt : discretized time step
        '''
        self.l = l
        self.m = m
        self.g = 9.81
        self.dt = dt
        
    def reset_env(self, state, epi_t):
        '''
        This function resets the state
        Input:
            state : [theta, theta_dot]
            epi_t : episode time
        '''
        
        self.sim_data = np.zeros((2, int(epi_t/self.dt)))
        self.sim_data[:,0] = state
        self.t = 0
        
    def compute_dynamics(self, theta, theta_dot, tau):
        '''
        This function returns xd = f(x)
        '''
        return np.array([theta_dot, (tau/(self.m*(self.l**2)))-(self.g/self.l)*np.sin(theta)])
     
    def rk4_integrate(self, theta, theta_dot, tau):
        '''
        ranga kutta integration of the step
        '''
        k1 = self.compute_dynamics(theta, theta_dot, tau)
        k2 = self.compute_dynamics(theta + 0.5*self.dt*k1[0], theta_dot + 0.5*self.dt*k1[1], tau)
        k3 = self.compute_dynamics(theta + 0.5*self.dt*k2[0], theta_dot + 0.5*self.dt*k2[1], tau)
        k4 = self.compute_dynamics(theta + self.dt*k3[0], theta_dot + self.dt*k3[1], tau)

        update_theta = (1/6)*self.dt*(k1[0] + 2*k2[0] + 2*k3[0] + k4[0])
        update_theta_dot = (1/6)*self.dt*(k1[1] + 2*k2[1] + 2*k3[1] + k4[1])

        return np.array([theta + update_theta, theta_dot + update_theta_dot])
        
    def integrate_dynamics(self, tau, horizon_length):
        '''
        integrates one step
        Input:
            tau : torque applied at current time step
            horizon_length : number of steps to be integrated
        '''            
        for i in range(int(horizon_length)-2):
            
            self.sim_data[0, self.t + 1], self.sim_data[1, self.t + 1] = \
                            self.rk4_integrate(self.sim_data[0, self.t], self.sim_data[1, self.t], tau[i]) 
                                                    
            self.t += 1
        
    def show_simulation(self, freq):
        '''
        This function shows the simulation of the pendulum for the given control
        Input:
            freq : frame rate
        '''
        sim_data = np.array(self.sim_data[:,::freq])
        fig = plt.figure()
        ax = plt.axes(xlim=(-self.l - 1, self.l + 1), ylim=(-self.l - 1 , self.l + 1))
        line, = ax.plot([], [], lw=3)
        dot1, = ax.plot([], [], 'o', color='red')
        dot2, = ax.plot([], [], 'o', color='blue')
        
        def init():
            line.set_data([], [])
            dot1.set_data([], [])
            dot2.set_data([], [])
            
            return line, dot1, dot2, 
        
        def animate(i):
            x = self.l*np.sin(sim_data[:,i][0])
            y = -self.l*np.cos(sim_data[:,i][0])
            line.set_data([0, x], [0, y])
            dot1.set_data([x], [y])
            dot2.set_data([0], [0])
            return line, 
        
        text_str = "pendulum"
        props = dict(boxstyle='round', facecolor='wheat', alpha=0.5)        
        anim = FuncAnimation(fig, animate, init_func=init,
                                       frames=np.shape(sim_data)[1], interval=25, blit=True)
        ax.text(0.05, 0.95, text_str, transform=ax.transAxes, fontsize=15,
        verticalalignment='top', bbox=props)
        plt.grid()
        plt.close(fig)
        plt.close(anim._fig)
        IPython.display.display_html(IPython.core.display.HTML(anim.to_html5_video()))
     
```

```python
x0 = torch.tensor([0.0, 0.0], dtype = float)
epi_t = 10
pend_env = PendulumEnv(1, 1, 0.01)
tau = 5.0*np.ones(int(epi_t/pend_env.dt) + 1)

pend_env.reset_env(x0, epi_t)
pend_env.integrate_dynamics(tau, epi_t/pend_env.dt)
pend_env.show_simulation(20)
```

```python
# this block contains functions to create the constraints

def compute_dynamics(theta, theta_dot, tau):
    '''
    This function returns xd = f(x)
    '''
    return torch.tensor([theta_dot, (tau -(9.81)*torch.sin(theta))], dtype = float)
    

def rk4_integrate(theta, theta_dot, tau, dt):
    '''
    ranga kutta integration of the step
    '''
    k1 = compute_dynamics(theta, theta_dot, tau)
    k2 = compute_dynamics(theta + 0.5*dt*k1[0], theta_dot + 0.5*dt*k1[1], tau)
    k3 = compute_dynamics(theta + 0.5*dt*k2[0], theta_dot + 0.5*dt*k2[1], tau)
    k4 = compute_dynamics(theta + dt*k3[0], theta_dot + dt*k3[1], tau)

    update_theta = (1/6)*dt*(k1[0] + 2*k2[0] + 2*k3[0] + k4[0])
    update_theta_dot = (1/6)*dt*(k1[1] + 2*k2[1] + 2*k3[1] + k4[1])

    return torch.tensor([theta + update_theta, theta_dot + update_theta_dot], dtype = float)

# def create_lambda(fun, t):
#     return lambda x : fun(x, t)

# def ce_11(x, t):
#     return x[t + 3] - rk4_integrate(x[t], x[t + 1], x[t + 2], 0.1)[0]
# def ce_12(x, t):
#     return x[t + 4] - rk4_integrate(x[t], x[t + 1], x[t + 2], 0.1)[1]
# def ci_11(x, t):
#     return x[t + 2] - 5.0 
# def ci_12(x, t):
#     return 5.0 - x[t + 2]

# def create_constraints_arr(horizon, dt):
#     ce = []
#     ci = []
#     ce.append(lambda x : x[0] - 2.5)
#     ce.append(lambda x : x[1])
#     for t in range(0, 3*(horizon), 3):
#         ce.append(create_lambda(ce_11, t))
#         ce.append(create_lambda(ce_12, t))
#         ci.append(create_lambda(ci_11, t))
#         ci.append(create_lambda(ci_12, t))
        
#     return ce, ci
        
# def cost(x):
#     return 10*x[-1]**2 + 10*(x[-2] - 3.14)**2
```

```python
u = 0

def ce01(x):
    return x[0] - 2.5
def ce02(x):
    return x[1] - 0.0
def ce11(x):
    return x[3] - rk4_integrate(x[0], x[1], x[2], 0.1)[0]
def ce12(x):
    return x[4] - rk4_integrate(x[0], x[1], x[2], 0.1)[1]
def ce21(x):
    return x[6] - rk4_integrate(x[3], x[4], x[5], 0.1)[0]
def ce22(x):
    return x[7] - rk4_integrate(x[3], x[4], x[5], 0.1)[1]
# def ce31(x):
#     return x[9] - rk4_integrate(x[6], x[7], x[8], 0.1)[0]
# def ce32(x):
#     return x[10] - rk4_integrate(x[6], x[7], x[8], 0.1)[1]

def ci01(x):
    return x[2] - u
def ci02(x):
    return u - x[2]
def ci11(x):
    return x[5] - u
def ci12(x):
    return u - x[5]
# def ci21(x):
#     return x[8] - u
# def ci22(x):
#     return u - x[8]

def cost(x):
    return x[7]**2 + (x[6] - 3.14)**2

ce = [ce01, ce02, ce11, ce12, ce21, ce22]
ci = [ci01, ci02, ci11, ci12]
```

```python
# This block to compute optimal control input
# control input bounded from -0.5 to 0.5
# dt = 0.1
# horizon = 2 # this is equal to horizon *dt seconds of planning
# n = 3*(horizon + 1) - 1
# ce, ci = create_constraints_arr(horizon, dt)
# print("Number of variables : " + str(n))
# print("Number of equality constraints : " + str(len(ce)))
# print("Number of inequality constraints : " + str(len(ci)))
bcf = BoundConstrainedFormulation(delta_max = 1.0, maxit = 500)
x_opt = bcf.optimize(cost, ce, ci, 8, 0.001, 0.001, 3, use_sr1 = False)
bcf.stats()
```

```python
print(cost(x_opt))
print(x_opt)
tau = []
theta = []
theta_dot = []
for i in range(0, len(x_opt) - 3, 3):
    theta.append((360*x_opt[i])/(2*np.pi))
    theta_dot.append(x_opt[i+1])
    tau.append(x_opt[i + 2])
plt.plot(theta)
plt.show()
```
